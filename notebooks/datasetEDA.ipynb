```python
from physionet2023.dataProcessing.cachedDataset import PhysionetPreprocessedDataset
from physionet2023.dataProcessing.TuhDatasets import TuhPreprocessedDataset

import matplotlib.pyplot as plt
import torch
import numpy as np
```

```python
pds = PhysionetPreprocessedDataset(path="../cache/physionet_cache")
tds = TuhPreprocessedDataset(path="../cache/tuh_cache")

print(len(pds))
print(len(tds))
```

```python
def draw_spectrogram(X: torch.Tensor):
    f = np.linspace(0.5, 30, X.shape[0])
    t = np.linspace(0, 5 * 60, X.shape[1])

    plt.ylabel('f [Hz]')
    plt.xlabel('t [sec]')
    plt.pcolormesh(t, f, X)
```

```python
X, y = pds[152]
draw_spectrogram(X[0,:])

print(X[0,:].mean())
print(X[0,:].std())
print(y)
```
