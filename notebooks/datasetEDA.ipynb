```python
from physionet2023.dataProcessing.datasets import PatientDataset
```

```python
pd = PatientDataset(root_folder="../data")
len(pd)
```

```python
pmd, rmd, r = pd[0]
pmd
```

```python
cpcs = [pmd['CPC'] for pmd, _, _ in pd]
cpcs
```

```python
frequency = {}

# iterating over the list
for item in cpcs:
   # checking the element in dictionary
   if item in frequency:
      # incrementing the counr
      frequency[item] += 1
   else:
      # initializing the count
      frequency[item] = 1

# printing the frequency
print(frequency)
```

```python
pt_count = sum([v for k, v in frequency.items()])
pt_count
```

```python
{k: (v / pt_count) for k, v in frequency.items()}
```
