```python
from physionet2023.dataProcessing.datasets import SampleDataset
import seaborn as sns
```

```python
ds_original = SampleDataset("../data", sample_len=1000, shuffle=True)
eeg_data, static_data, label = ds_original[0]
sns.lineplot(x=range(0, eeg_data.shape[-1]), y=eeg_data[0,:])
```

```python
RESAMPLE_FACTOR = 5

ds_resampled = SampleDataset("../data", sample_len=int(1000 / RESAMPLE_FACTOR), shuffle=True, resample_factor=RESAMPLE_FACTOR)
eeg_data, static_data, label = ds_resampled[0]
sns.lineplot(x=range(0, eeg_data.shape[-1]), y=eeg_data[0,:])
```
